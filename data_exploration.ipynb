```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join('data','heart_failure_clinical_records_dataset.csv'))
df.head()
```

```python
for i in df.columns:
    sns.distplot(df[f"{i}"])
    plt.show()
```
